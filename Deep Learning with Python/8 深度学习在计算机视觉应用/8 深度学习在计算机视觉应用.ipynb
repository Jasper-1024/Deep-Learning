# 8 深度学习在计算机视觉应用

章节概况

- 了解卷积神经网络 cnn
- 使用数据增强降低过拟合
- 使用预训练的卷积神经网络进行特征提取
- 微调预训练的卷积神经网络

神经网络在计算机视觉的成功应用,是深度学习重新走入人们视野的开始,历史这一段就不细究了.

本章主要介绍了卷积神经网络,卷积神经网络是另一种类型的神经网络模型,在计算机视觉领域有着广泛的应用.本章会涉及 cnn 应用在图像分类问题--特别是小型数据集.


## 8.4 章节概括

卷积神经网络是计算机视觉最好的机器学习模型,即使是非常小的数据集也能有不错的效果.

图像数据特有的数据增强,对抗小数据集的过拟合.

特征提取复用预处理的模型,在小数据集异常管用.

特征提取+模型微调,是现在在小数据解决图像分类问题的最终办法.(目前)
