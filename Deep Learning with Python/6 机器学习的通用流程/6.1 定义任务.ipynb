# 定义任务

如果不了解将要进行工作的背景,别想顺利完成工作.我们需要知道

- 为什么客户想解决这个问题?
- 问题解决,他们的收益是什么?
- 你的模型,客户想如何使用它?
- 可以使用的数据有哪些?
- 最后,这样的问题可以映射成为怎样的机器学习任务?

## 定义问题

现实世界往往模糊不堪,一个问题千人千面.和利益相关者深入探讨完整定义问题是机器学习的第一步.

考虑下面几个问题

- 你的输入数据是什么?
- 面临的是那种类型的机器学习任务?
- 现有的解决方案是什么?
- 特别的限制条件?

数据: 有数据才有一切,有有用的数据才有模型.数据可用性是这个阶段首要考虑的问题,下一节我们会详细介绍.

机器学习类型:

- 机器学习任务本身有非常多分类,二分类/多标签多分类/标量回归/矢量回归/图像分割/聚类/强化学习...
- 机器学习不是万能的,有许多经典问题已经存在更好的解法.
- 例子
  - 照片分类,多标签多分类问题
  - 垃圾邮件检测.二分类问题
  - 音乐推荐,最佳解决方案不是机器学习,而是需要矩阵分解等等处理.
  - 信用卡欺诈检查,二分类问题
  - 特定图像检索,需要收集特征图片,再训练.

已有解决方案

- 或许客户已经有现成的解决方案运行,了解它并最终超越它.

特别的限制因素

- 现实世界总有各种限制,在解决问题时需要了解清楚
- 数据几乎是端到端加密?模型运行在嵌入式设备?等等

一旦回答了上面的问题,此时应该已经了解了

- 输入模型的数据
- 最终的目标
- 属于那种机器学习类型
- 提出有关问题的假设.

最后提出的假设,只是现阶段的假设而已.必须在后面的阶段证明或者证伪.


## 构建数据集

详细定义了问题后,我们来到了机器学习项目中通常最耗时,最艰巨,最昂贵的部分 --> 数据集.

足够可用的数据集,是模型的几乎全部基础,如果你有额外的资源,那么直接投入到数据上.

确定需要收集的数据

- 数据千千万,但是与你的问题相关的数据才有价值.
- 实际项目的各种限制条件.

数据标注的基础设施

- 数据标注的质量决定了数据集的质量.
- 数据标注是非常繁重的任务,合适的自动化会提高效率.
- 考虑外包,但是可能会失去标注的质量.
- 专业化的数据标签公司,更好的质量,更昂贵的价格.

选择最佳方案,考虑工作的限制.

- 标注所需要的最低基础,是否任何人都可以进行标注?
- 标注需要训练,那如何训练进行标注的人?
- 你自己是否了解标注的过程,相关领域专家又是如何进行标注的,你能否理解?
- 自行标注的工具

警惕非代表性数据

一个著名的例子是 1948 美国大选预测,芝加哥报社进行非常大规模的问卷调查,样本数量超过了以往,结果非常离谱的错了.他们的样本是打电话询问的,1948年电话尚未普及,家庭安装电话已经代表了一定经济实力.以这样的样本代替全部,结果自然离谱.

机器学习模型只能理解与数据相似的输入,无论如何要谨记.收集的数据不能代表全局,那欢迎来到机器学习的地狱...

如果可能,自己在生产环境中收集最直接的数据.对 imdb 影评分级,直接爬最新的 imdb 评论.对食物打分,直接获取人们分享的晚餐.

注意概念漂移,随着时间推移,任何概念都在发生变换.一个在 2011 年工作很好的微博年度情感统计,在2021年应该会几乎不可用.太多的沧海桑田,太多语义的变动,用户群体变换,法律问题等等.

再重复一次,机器学习只能记忆过去输入数据的模式,使用过去的数据训练模型,是在假定未来规律本身没有发生变化,但是变化最快的往往是规律,现象反而会更加迟缓.特别是与人相关的领域.


## 理解数据本身

将收集到的数据作为黑箱直接扔给机器学习,是非常不可取的行为.开始训练前,我们需要探索收集的数据.

数据包含文本/图像,直接查看样本.

数据含有数字,尝试了解取值范围,频率等等信息.可视化会比干巴的统计数据更有效.

数据包含位置信息,汇总在地图上.一目了然总会找到一些规律.

是否会有特征值缺失?分布异常等等(下一节会谈到异常数据处理)

如果是分类问题,样本时候均匀?如果不是,如何解释?是否是样本本身存在问题?

最重要的,确定没有信息泄露.假设模型的任务是预测癌症,结果医疗数据中存在,此人已患癌症.那这条记录是无效的,应该去除.


## 选择衡量标注

要成功解决问题,现在数据有了,回到成功解决问题本身.什么算是成功解决?需要精确定义.

如果是分类问题,那可能是正确率.

如果是广告推荐系统,可能是广告商的引流数据.

有时我们还需要自定义指标,这个指标要让利益相关者接受.
