# 机器学习的通用流程

本站的主要内容有(搁着新闻联播呢)

- 定义机器学习解决的问题
- 开发一个可以工作的机器学习模型
- 在生产环境部署模型,并进行定期维护.

我们之前的例子都是已经有了数据集就等训练,但是现实世界往往不同.现实世界往往是从问题定义开始的.

通常机器学习的工作流程有 3 部分

- 定义任务: 了解问题所在领域,了解客户要求,收集数据集,了解数据,选择衡量任务的指标.
- 开发模型: 选择模型评估协议,准备一个非机器学习的基线指标.调整模型超参数,训练模型平衡过拟合/欠拟合,正则化等 MAX 泛化.
- 部署模型: 向利益相关者介绍模型,最后将模型部署到生产环境.同时要监视生产环境,收集更新下一代模型的数据.

## 人工智能的道德问题

这是文章的一个小贴士部分,但是值得单独拿出来说.

特别是算法日益嵌入生活的现在.受到算法执行结果影响的将是活生生的人.而且很有可能不是几张优惠券,而是工作/贷款/信用甚至是法律问题.

作者的观点是,技术从来就不是中立的,当你的工作可能对现实世界产生如此的影响时,这种技术选择也成为了道德选择.希望你能谨慎的考虑你的工作所支持的价值观.
