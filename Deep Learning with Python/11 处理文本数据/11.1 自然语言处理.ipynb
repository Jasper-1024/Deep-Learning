11.1 自然语言处理

计算机科学中,将人类的语言(英语,普通话)称为自然语言(natural languages),以区别机器语言/编程语言.

每一个编程语言的起点都是人类工程师写下一套规则,来描述如何如何使用这些语言/语句.这意味对编程语言,规则是第一位的.现有规则才有语句.

然而自然语言是完全相反的,自然语言是使用在前,规则产生在后.犹如生物进化.自然语言的规则是事后被确定的,这意味使用者经常会打破规则.

因此,编程语言是高度结构化的,使用精确规则将固定词汇表达非常精确的概念.但是自然语言完全相反,模糊/混乱/不断变化.

首先是我们的一切几乎都是建筑在自然语言之上的,大部分的交流,我们的文化,互联网大部分也是文本,语言承载了几乎所有知识,我们的意识思考也是建立在语言之上的.然而理解自然语言的能力一直是机器难以企及的.最开始人们是已于规则的自然语言处理,比如写下英语的规则集,进行简单的机器翻译/聊天机器人.但是语言本身非常叛逆,非常不容易形式化,经过很长时间的发展,这些系统依旧效果不佳.

人工编写规则一直主导到了上世纪 90 年代,从上世纪 80 年代开始,计算机和数据开始提供一个更好的方案.让计算机从一堆语料库中自动寻找一些模式,这本身是机器学习的一部分了.80年代末,最早是基于决策树的方法.后来统计学方法开始回归,随着时间推移学习型参数模型完全占据主体,早期语言识别专家 弗雷德里克-耶利内克（Frederick Jelinek）曾开玩笑说:'每当解雇一个语言学家,语音识别的性能就会上升'.

这就是现代 nlp 的意义,不是理解语言,只是利用机器学习和大型数据集,试图返回一些有用的东西,例如预测等等.

- 文字的主题? -> 文本分类
- 文字包含抄袭/滥用 -> 文本过滤
- 这段文字是积极还是消极 -> 情感分析
- 这段不完整的话语,下一个词是什么? -> 语言建模
- 这句话用德语怎么说? -> 机器翻译
- 如何概况这篇文章? -> 总结

无论如何要记住,现有的 nlp 并不是尝试理解语言,只是从数据中统计规律.事实上这就已经足够在很多任务中表现出色.

nlp 的工具从上世纪 90 年代到 2010 年代只经历了缓慢的演变,在 2014 年以前 nlp 的工具还是决策树/逻辑回归,大多数只是特征工程的研究.但是在 2014-2015 年以后研究人员开始研究 rnn 对 nlp 的效果,特别是 ltsm.

2015 年初,keras 提供了 ltsm 的实现,然后的 2015-2017 年 rnn 主导了 nlp 的发展.最后 2017-2018 年一个新的架构取代了 rnn,这就是 Transformer,现如今的各类 nlp 大多都是基于 Transformer,我们将在章节的后半部分聊到.
