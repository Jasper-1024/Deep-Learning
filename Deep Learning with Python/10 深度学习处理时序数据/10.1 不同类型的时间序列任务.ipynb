## 10.1 不同类型的实践序列数据

时间序列顾名思义,是与时间耦合的数据可以是任意定期测量得到的数据.典型的时间序列数据: 股票 商店的销售额 自然行为(天气 地震等等).

时间序列数据与前文我们接触到的数据都非常不同,使用时间序列数据涉及到对系统动态的理解.包括周期性循环,随时间变化的趋势,还有突然性尖峰.

最基本的时间序列任务是预测,预测接下来数据将会如何变化.预测将来用电量,预测天气等等.但是预测只是一个任务,

- 分类: 查看时间序列数据,划分 1 个或多个标签.例如: 查看网站的行为数据,标记人还是机器人.(题外话,现在机器人比人还真,我基本上每天都会被识别成机器人)
- 事件检测: 从一个连续的数据流中识别/预期特定的事件.一个特别的应用是热词检测,监控音频流,检查 ok google 这样的词语.
- 异常检测: 检查连续数据流的异常情况.例如: 向模型灌入生产环境的很长时间的数据,如果有外来入侵者,某些数据会产生变化,从而检测异常.当一开始时,人是不知道在寻找什么,一般是无监督学习.

处理时序数据在信息领域已经有非常多的方法了,例如傅里叶变换.在深度学习的背景下,傅里叶变换等其他领域的技术可以当作特征工程使用,使得模型更加容易工作.然而本章内容重点是模型,因此不会有过多涉及.

本章主要内容是 RNN 和将其应用在时间序列任务上.
