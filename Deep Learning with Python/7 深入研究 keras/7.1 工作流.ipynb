# 7.1 工作流

Keras API 设计是以复杂度逐渐显现为原则的:

- 让新手容易入门,又能足够完成处理高度复杂的任务,只需要一步一步的循序渐进学习.
- 简单用例的实现应该足够简单,但是实现任意的高级用例也是可能的,无论你向做的事情有多小众,都会有明确的路径.

这意味着,从新手到专家,我们依旧在使用相同的工具,只是方式会非常不一样.

keras 种没有真正的方法(个人理解是固定死的 api),相反 keras 提供了一系列工作流,从非常简单到非常灵活复杂,有许多不同的路径来构建 keras 模型,也有许多不同方式训练模型.所有的这些工作流都是基于相同的共享 api,例如 Layer 和 Model.不同工作流的组件可以互相自由组合.
